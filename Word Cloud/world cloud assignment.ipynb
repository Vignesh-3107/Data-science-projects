```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('World cloud.csv')
```

```python
txt=df['t']
```

```python
data=''
for i in txt:
    data=data+i
```

```python
for i in ".,:;!@#$%^&*/\?<->')]}[1234567890{(`":
    data=data.replace(i,' ')
```

```python
words = data.split(' ')
print("Total words in given data : ", len(words))
print("Unique words in given data : ",len(list(set(words))))

Unique_words = list(set(words))
```

```python
word_list = {}
count = 0
for i in Unique_words:
    word_list[i] = count
    count += 1
```
